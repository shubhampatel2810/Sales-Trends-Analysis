# Now we See Visualization 

```python
plt.figure(figsize=(10, 6))
sns.histplot(doc['Age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
doc.head(2)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Total Amount', data=doc)
plt.title('Scatter Plot of Age vs. Total Amount')
plt.xlabel('Age')
plt.ylabel('Toatl Amount')
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.countplot(x='Product Category', data=doc)
plt.title('Count of Each Product Category')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.boxplot(x='Product Category', y='Age', data=doc)
plt.title('Box Plot of Value by Age')
plt.xlabel('Product Category')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(10, 3))
sns.violinplot(x='Age', y='Total Amount', data=doc)
plt.title('Vilin Plot chart for spend highest money age')
plt.xlabel('Age')
plt.ylabel('Toatl Amount')
plt.show()
```

```python
plt.figure(figsize=(10, 3))
sns.countplot(x='Age', data=doc)
plt.title('Number of customer by age  ')
plt.xlabel('Age')
# plt.ylabel('Toatl Amount')
plt.show()
```

```python
plt.figure(figsize=(8, 4))  # Set the figure size
sns.countplot(x='Date', data=doc, order=doc['Date'].value_counts().index[:10])  # Order by the top 10 dates
plt.title('Top 10 Dates with the Highest Sales')
plt.xlabel('Date')
plt.ylabel('Number of Sales')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```
